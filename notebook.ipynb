```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
plt.style.use('ggplot')
import seaborn as sns
```

```python
netflix_df['type'].value_counts().plot(kind='bar' ,  color=['magenta' , 'tab:cyan' ,])
plt.show()
```

```python
netflix_df['genre'].value_counts()
```

The most frequent movie duration in the 1990s

```python
netflix_movies_90s= netflix_df[ (netflix_df['type']=='Movie') & (netflix_df['release_year']>=1990) 
&(netflix_df['release_year']<2000)]
netflix_movies_90s
```

```python
duration =int(netflix_movies_90s['duration'].mode())
print('duration=', duration)
```

```python
plt.hist(netflix_movies_90s['duration'], bins=20, color='tab:cyan')
plt.xlabel('Duration (in minutes)' , color='magenta')
plt.ylabel('Number of Movies' , color='magenta')
plt.title('Histogram of Movie Durations in 1990s' , color='lime' , y=1.03)
plt.show()
```

```python
#Number of movies per duration
print(netflix_movies_90s['duration'].value_counts(sort=True))
```

Number of Short Movies of 1990s

```python
short_movie_count= len(netflix_df[(netflix_df['duration']<90) & (netflix_df['genre']=='Action')
& (netflix_df['type']=='Movie') & (netflix_df['release_year']>=1990) & (netflix_df['release_year']<2000)])
print(short_movie_count)
```

```python
#Plot of Short Movies of 1990s with their respective durations
short_action_movies=netflix_df[(netflix_df['duration']<90) & (netflix_df['genre']=='Action')
& (netflix_df['type']=='Movie') & (netflix_df['release_year']>=1990) & (netflix_df['release_year']<2000)] 
sns.catplot(y='duration'  , x='title' , data =short_action_movies , kind='bar' , ci=None ) 
plt.xticks(rotation=90)
plt.title('Short Movies of 1990s', color='lime' , y=1.03)
plt.show()
```
